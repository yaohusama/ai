```python
import pandas as pd
import pandas as pd
import numpy as np
import seaborn as sns #热力图
import itertools 
import datetime
import matplotlib.pyplot as plt
import statsmodels.api as sm 
from statsmodels.tsa.stattools import adfuller #ADF检验
from statsmodels.stats.diagnostic import acorr_ljungbox #白噪声检验
from statsmodels.graphics.tsaplots import plot_acf,plot_pacf #画图定阶
from statsmodels.tsa.arima_model import ARIMA #模型
from statsmodels.tsa.arima_model import ARMA #模型
from statsmodels.stats.stattools import durbin_watson #DW检验
from statsmodels.graphics.api import qqplot #qq图
train=pd.read_csv("D:/mathcup/traindataset/train181.csv",parse_dates=[['Date','time']])
print(train.columns)
train.index=train.Date_time
train.head(10)
```

```python
diff1=train['upGB'].diff(1).dropna()
plt.figure()
diff1.plot(color = 'red',title='diff 1',figsize=(15,10))
plt.show()
diff2 = diff1.diff(1)
diff2.plot(color = 'b',title='diff 2',figsize=(15,10))
```

```python
#白噪声检验
p_value = acorr_ljungbox(train['upGB'], lags=1) 
print (p_value)
```

```python
plot_acf(train['upGB'],lags=40) #延迟数
plot_pacf(train['upGB'],lags=40)
plt.show()
```

```python
p_min = 0
q_min = 0
p_max = 5
q_max = 5
d_min = 0
d_max = 5
# 创建Dataframe,以BIC准则
results_aic = pd.DataFrame(index=['AR{}'.format(i) \
                            for i in range(p_min,p_max+1)],\
        columns=['MA{}'.format(i) for i in range(q_min,q_max+1)])
    # itertools.product 返回p,q中的元素的笛卡尔积的元组
for p,d,q in itertools.product(range(p_min,p_max+1),\
                                range(d_min,d_max+1),range(q_min,q_max+1)):
    if p==0 and q==0:
        results_aic.loc['AR{}'.format(p), 'MA{}'.format(q)] = np.nan
        continue
    try:
        model = sm.tsa.ARIMA(train['upGB'], order=(p, d, q))
        results = model.fit()
        #返回不同pq下的model的BIC值
        results_aic.loc['AR{}'.format(p), 'MA{}'.format(q)] = results.aic
    except:
        continue
results_aic = results_aic[results_aic.columns].astype(float)
#print(results_bic)
    
fig, ax = plt.subplots(figsize=(10, 8))
ax = sns.heatmap(results_aic,
                #mask=results_aic.isnull(),
                ax=ax,
                annot=True, #将数字显示在热力图上
                fmt='.2f',
                )
ax.set_title('AIC')
plt.show() 
```

```python
#建立模型
arma_model = ARMA(train['upGB'],order=(1,3))
result = arma_model.fit()#激活模型
pred_one = result.predict(start=len(train)-5,end = len(train)+ 168,dynamic=True)
pred_one
```
